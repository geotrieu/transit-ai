```python
import pandas as pd
import plotly.express as px
```

```python
stations = getSeoulStations()
```

```python
fig = px.scatter_mapbox(stations, 
                        lat="lat", 
                        lon="lng", 
                        hover_name="name",
                        zoom=8, 
                        height=800,
                        width=800)

fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
def getSeoulStations():
    df = pd.read_csv('../res/datasets/csv/seoul_subway_stations.csv', sep=',', index_col=0);
    df = df.drop(['line', 'no'], axis=1);
    return df;
```
